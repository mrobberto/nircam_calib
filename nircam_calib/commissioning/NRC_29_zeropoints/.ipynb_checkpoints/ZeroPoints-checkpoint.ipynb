## F070W

```python
import os
from astropy.utils.data import get_pkg_data_filename
from synphot import SpectralElement
from synphot.models import Empirical1D
bp = SpectralElement.from_file(F070W_file)
bp.waveset.to(u.micron)
#convoluted way to change the wl units from micron to A. There must be an easier way!
wl = bp.waveset * 1E4
th = bp.model.lookup_table
F070W = SpectralElement(Empirical1D, points=wl, lookup_table=th, keep_neg=True)
```

```python
F070W.plot(left=1000, right=10000)
```

```python
from synphot import Observation,units
area = 25*1E4 * units.AREA
binset = range(5000, 10001)

obs_P330e_F070W = Observation(sp, F070W, binset=binset)
binflux = obs_P330e_F070W.sample_binned(flux_unit='count', area=area)
# Sample the "native" flux for comparison
flux = obs_P330e_F070W(binset, flux_unit='count', area=area)
flux_sum = flux.sum()
print(flux_sum)
```

```python
obs_vega_F070W = Observation(vega, F070W, binset=binset)
binflux = obs_vega_F220W.sample_binned(flux_unit='count', area=area)
# Sample the "native" flux for comparison
flux_Vega = obs_vega_F070W(binset, flux_unit='count', area=area)
flux_Vega_sum = flux_Vega.sum()
```

```python
import numpy as np
mag_P330e_F070W = -2.5 * np.log10(flux_sum/flux_Vega_sum)
print(mag_P330e_F070W)
```
